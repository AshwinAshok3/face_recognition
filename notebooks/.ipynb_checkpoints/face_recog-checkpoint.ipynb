```python

import cv2
import numpy as np
import insightface
from tqdm import tqdm
import os
from pathlib import Path
from numpy.linalg import norm
```

```python
# =============================
# 2. Initialize Model (Auto GPU/CPU Detection)
# =============================
try:
    app = insightface.app.FaceAnalysis(name='buffalo_l', providers=['CUDAExecutionProvider', 'CPUExecutionProvider'])
except Exception as e:
    print(f"GPU not available or driver issue: {e}\nFalling back to CPU mode.")
    app = insightface.app.FaceAnalysis(name='buffalo_l', providers=['CPUExecutionProvider'])

app.prepare(ctx_id=0, det_size=(640, 640))
```

```python
# +os.getcwd()
```

```python

# =============================
# 3. Load Reference Faces
# =============================
reference_folder = Path('reference_faces')
if not reference_folder.exists():
    raise FileNotFoundError("Folder 'reference_faces' not found. Please create and add known faces.")

known_faces = []
known_names = []

for image_path in reference_folder.glob('*.jpg'):
    img = cv2.imread(str(image_path))
    faces = app.get(img)
    if len(faces) == 0:
        print(f"âš ï¸ No face detected in {image_path.name}")
        continue
    known_faces.append(faces[0].embedding)
    known_names.append(image_path.stem)

if len(known_faces) == 0:
    raise Exception("No valid reference faces found.")

known_faces = np.array(known_faces)
```

```python







# =============================
# 4. Utility: Cosine Similarity
# =============================
def cosine_similarity(a, b):
    return np.dot(a, b) / (norm(a) * norm(b))


def find_best_match(face_embedding, threshold=0.35):
    similarities = [cosine_similarity(face_embedding, known_face) for known_face in known_faces]
    best_match_idx = np.argmax(similarities)
    best_score = similarities[best_match_idx]
    if best_score > threshold:
        return known_names[best_match_idx], best_score
    else:
        return "Unknown", best_score

# =============================
# 5. Process Video Feed (MP4 or Webcam)
# =============================
video_source = 0  # Use 0 for webcam, or path like 'crowd_scene.mp4'
cap = cv2.VideoCapture(video_source)

if not cap.isOpened():
    raise Exception("Cannot open video source. Check your camera or file path.")

print("\nâœ… Face Recognition System Initialized. Press 'q' to quit.\n")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    faces = app.get(frame)
    for face in faces:
        bbox = face.bbox.astype(int)
        embedding = face.embedding
        name, score = find_best_match(embedding)

        color = (0, 255, 0) if name != "Unknown" else (0, 0, 255)
        cv2.rectangle(frame, (bbox[0], bbox[1]), (bbox[2], bbox[3]), color, 2)
        cv2.putText(frame, f"{name} ({score:.2f})", (bbox[0], bbox[1] - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    cv2.imshow('InsightFace Recognition', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# =============================
# 6. Notes for Stability under Jupyter + WSL2
# =============================
# - Always use cv2.imshow() in a separate window (avoid inline plotting).
# - Avoid matplotlib live loops; they can cause Jupyter kernel death in WSL.
# - If using VSCode Jupyter, ensure your kernel uses the same Conda env (face_env).
# - GPU fallback is automatic; you donâ€™t need to change providers manually.

print("\nâœ… Execution complete. System stable for WSL2 Conda Jupyter Notebook.")
```
